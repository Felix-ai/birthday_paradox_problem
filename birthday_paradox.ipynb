```python
# Import Modules
from math import factorial
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use("ggplot")
```

```python
# Parameters
days = 366
p = 0
val = 0
ps = []
vals = []
```

```python
# Define Function
def prob_shared_birthday(d, N):
    return 1 - factorial(d) / (factorial(d-N) * d**N)
```

```python
# Score probabilities for different group sizes
while p >= 0 and p <= 1:
    p = prob_shared_birthday(days, val)
    val+=1
    ps.append(p)
    vals.append(val)
    if p == 1:
        break
```

```python
# Plot distribution
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(vals, ps, color="g")
ax.set_xlabel("Group Size", color="b")
ax.set_ylabel("P(shared birthdays)", color="b")
ax.set_title("Birthday Paradox", color="k", fontsize=14)
plt.savefig("images/birthday_paradox_solution.svg")
```
